```python
import numpy as np
from scipy.special import comb
from matplotlib import pyplot as plt
```

# Hyperexponential Distribution Function

```python
# Parameters:
#    lambs: a list of values of lambda
#    alphas: a list of values of alpha
#    t: time to the first success
#    res: resolution
# Returns:
#    [cdf, pdf, res, t]
def hyperexponential(lambs, alphas, t, res):
    x = np.linspace(0.2, t, res)
    pdf = 0
    cdf = 0
    
    for i in range(len(lambs)):
        pdf = pdf + alphas[i]*lambs[i]*np.exp(-lambs[i]*x)
        cdf = cdf + alphas[i]*(1 - np.exp(-lambs[i]*x))
    
    return [cdf, pdf, res, t]
```

Test Hyperexponential Distribution function

```python
hyperexponential([2, 1], [0.3, 0.7], 7, 50)
```

# Plot the continuous ditribution

```python

# Plotting continuous distribution function 
def continuous_dist_plot(dist_name, dist_vals):
    cdf = dist_vals[0]
    pdf = dist_vals[1]
    res = dist_vals[2]
    n = dist_vals[3]
    
    x = np.linspace(0.2, len(cdf)*n/res, res)
    
    # Plot cdf
    plt.figure(0)
    plt.plot(x, cdf, color='k', lw = 1.2)
    plt.title('CDF of ' + dist_name.capitalize() + ' Distribution')
    plt.ylabel('F(k)')
    plt.xlabel('k')
    plt.xlim([0, 9])
    plt.ylim([0, max(cdf) + 0.2])
    plt.grid(which = 'both', linestyle='--', axis = 'y')
    plt.savefig('cdf_' + dist_name + '.png', dpi=150)
    
    # Plot pdf
    plt.figure(1)
    plt.plot(x, pdf, color='k', lw = 1.2)
    plt.title('PDF of ' + dist_name.capitalize() + ' Distribution')
    plt.ylabel('f(k)')
    plt.xlabel('k')
    plt.xlim([0, 9])
    plt.ylim([0, max(pdf) + 0.1])
    plt.grid(which = 'both', linestyle='--', axis = 'y')
    plt.savefig('pdf_' + dist_name + '.png', dpi=150)
    
    plt.show()
```

```python
continuous_dist_plot('hyperexponential', hyperexponential([2, 1], [0.3, 0.7], 7, 50))
```
